```python
import imagine as im
import json
import pandas as pd
```

```python
data_dir = im.config.EXTERNAL_DATA_DIR 
models = [
    "gemma3:12b",
    "gemma3:12b-it-qat",
    "gemma3:27b",
    "gemma3:27b-it-qat",
    "llama3.3:70b",
    "llama4:scout",
]
tasks = [
    "vviq_independent_item",
    "vviq_cumulative_item",
    "psiq_cumulative_item",
    "psiq_independent_item"
]
```

```python
all_data = []
z_st = pd.DataFrame()
z_co = pd.DataFrame()

for task in tasks:
    for model in models:
        if "independent" in task:
            task_dir = data_dir / task / f"ollama_{model}_SingleTurn"
        else:
            task_dir = data_dir / task / f"ollama_{model}_Convo"
        
        for sim in task_dir.iterdir():
            if ".psyscan" in sim.name:
                with open(sim,"r") as f:
                    x = json.load(f)
                    x = pd.DataFrame(im.sim_data.sim_data_extract(x))
                    x["model"] = model
                    x["task"] = task
                    if "independent" in task:
                        z_st = pd.concat([z_st,x])

                    if "cumulative" in task:
                        z_co = pd.concat([z_co, x])
        print(task_dir.is_dir(),task_dir)
```

```python
z_st["imagination_ability"] = z_st.apply(lambda x: im.sim_data.extract_population(x), axis=1)
z_co["imagination_ability"] = z_co.apply(lambda x: im.sim_data.extract_population(x), axis=1)
```

```python
vviq_i_data = z_st.loc[z_st["task"] == "vviq_independent_item"]
psiq_i_data = z_st.loc[z_st["task"] == "psiq_independent_item"]

vviq_c_data = z_co.loc[z_co["task"] == "vviq_cumulative_item"]
psiq_c_data = z_co.loc[z_co["task"] == "psiq_cumulative_item"]
```

```python
col_vviq_map = {
    'A_1' : 'A_1',
    'A_2' : 'A_2',
    'A_3' : 'A_3',
    'A_4' : 'A_4',
    'B_5' : 'B_1',
    'B_6' : 'B_2',
    'B_7' : 'B_3',
    'B_8' : 'B_4',
    'C_9' : 'C_1',
    'C_10' : 'C_2',
    'C_11' : 'C_3',
    'C_12' : 'C_4',
    'D_13' : 'D_1',
    'D_14' : 'D_2',
    'D_15' : 'D_3',
    'D_16' : 'D_4',
    'E_17' : 'E_1',
    'E_18' : 'E_2',
    'E_19' : 'E_3',
    'E_20' : 'E_4',
    'F_21' : 'F_1',
    'F_22' : 'F_2',
    'F_23' : 'F_3',
    'F_24' : 'F_4',
    'G_25' : 'G_1',
    'G_26' : 'G_2',
    'G_27' : 'G_3',
    'G_28' : 'G_4',
    'H_29' : 'H_1',
    'H_30' : 'H_2',
    'H_31' : 'H_3',
    'H_32' : 'H_4'
}

col_psiq_map = {
    'App_1' : 'App_1',
    'App_2' : 'App_2',
    'App_3' : 'App_3',
    'So_4' : 'So_1',
    'So_5' : 'So_2',
    'So_6' : 'So_3',
    'Sm_7' : 'Sm_1',
    'Sm_8' : 'Sm_2',
    'Sm_9' : 'Sm_3',
    'Ta_10' : 'Ta_1',
    'Ta_11' : 'Ta_2',
    'Ta_12' : 'Ta_3',
    'To_13' : 'To_1',
    'To_14' : 'To_2',
    'To_15' : 'To_3',
    'Bo_16' : 'Bo_1',
    'Bo_17' : 'Bo_2',
    'Bo_18' : 'Bo_3',
    'Fe_19' : 'Fe_1',
    'Fe_20' : 'Fe_2',
    'Fe_21' : 'Fe_3',
}
```

```python
vviq_i_data["trid"] = vviq_i_data["trid"].map(col_vviq_map)
vviq_c_data["trid"] = vviq_c_data["trid"].map(col_vviq_map)
psiq_i_data["trid"] = psiq_i_data["trid"].map(col_psiq_map)
psiq_c_data["trid"] = psiq_c_data["trid"].map(col_psiq_map)
```

```python
vviq_i_data2 = im.sim_data.pivot_ai_vividness(vviq_i_data)
psiq_i_data2 = im.sim_data.pivot_ai_vividness(psiq_i_data)
vviq_c_data2 = im.sim_data.pivot_ai_vividness(vviq_c_data)
psiq_c_data2 = im.sim_data.pivot_ai_vividness(psiq_c_data)
```

```python

vviq_i_data2["score"] = vviq_i_data2.sum(axis=True)
psiq_i_data2["score"] = psiq_i_data2.sum(axis=True)
vviq_c_data2["score"] = vviq_c_data2.sum(axis=True)
psiq_c_data2["score"] = psiq_c_data2.sum(axis=True)
```

```python
vviq_ai = pd.concat([vviq_i_data2,vviq_c_data2])
psiq_ai = pd.concat([psiq_i_data2,psiq_c_data2])
```
